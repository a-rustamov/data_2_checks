```python
import requests
import pandas as pd
import json
api_lit = "http://api.worldbank.org/v2/country/All/indicator/SE.ADT.LITR.ZS?format=json&date=2010:2020"
api_gdp = "http://api.worldbank.org/v2/country/All/indicator/NY.GDP.PCAP.CD?format=json&date=2010:2020"
```

```python
response_lit = requests.get(api_lit)
response_gdp = requests.get(api_gdp)
```

```python
#response_gdp.json()
```

```python
data_lit = response_lit.json()
data_gdp = response_gdp.json()
```

```python
#Normalaize nested json
df_lit= pd.json_normalize(data_lit[1])
df_gdp= pd.json_normalize(data_gdp[1])
#df_gdp
```

```python
#Remove rows with NaN in value
df_lit_cleaned = df_lit.dropna(subset='value')
df_gdp_cleaned = df_gdp.dropna(subset='value')
```

```python
#df_gdp_cleaned
```

```python
#Modify Date column type to integer for future queries
df_gdp_cleaned['date'] = df_gdp_cleaned['date'].astype('int64')
```

```python
avrg_lit = df_lit_cleaned["value"].mean()
print("Average literacy level for all adults around the world 2010-2020 \n", avrg_lit)
```

```python
num_cnt = len(pd.unique(df_lit_cleaned['country.id']))
print("Number of countries in the Literacy dataset \n", num_cnt)

```

```python
avrg_gdp = df_gdp_cleaned["value"].mean()
print("Average GDP in 2010-2020 \n", avrg_gdp)
```

```python
num_rgn = len(pd.unique(df_gdp_cleaned['country.value']))
print("Number of Regions in the GDP dataset \n", num_rgn)
```

```python
avg_lit_year = df_lit_cleaned.groupby('date')["value"].mean()
print("Average literacy level for all adults around the world by year \n", avg_lit_year)

```

```python
avrg_gdp_year = df_gdp_cleaned.groupby('date')["value"].mean()
print("Average GDP  by year \n", avrg_gdp_year)

```

```python
#Pandas query
lit_query = df_lit_cleaned.query("value > 80")
print("Literacy level over 80% \n", lit_query)
```

```python
gdp_query = df_gdp_cleaned.query("date == 2020 and value < 2000")
print("GDP less than 2000 for year 2020  \n", gdp_query)
```

```python
print(" Literacy 2nd and 3rd columns  \n", df_lit_cleaned[['countryiso3code', 'date']])
print(" GDP 2nd and 3rd columns  \n", df_gdp_cleaned[['countryiso3code', 'date']])
```
